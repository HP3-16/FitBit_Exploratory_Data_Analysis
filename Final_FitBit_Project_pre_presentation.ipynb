### **Categorising Heart Rate Ranges**

```python
def heart_reading(val):
  if val<=50:
    return "Athelete"
  elif val>50 and val<=70:
    return "Normal"
  elif val>70 and val<=100:
    return "Low Risk"
  elif val>100 and val<=150:
    return "Medium Risk"
  else:
    return "High Risk"
df_3['HeartStatus'] = df_3['Value'].apply(heart_reading)
```

```python

plt.figure(figsize=(15,5))
counts_1, bin_edges_1 = np.histogram(mild_activity['Calories'], bins=10, 
                                 density = True)

counts_2, bin_edges_2 = np.histogram(moderate_activity['Calories'], bins=10, 
                                 density = True)

counts_3, bin_edges_3 = np.histogram(high_activity['Calories'], bins=10, 
                                 density = True)
pdf_1 = counts_1/(sum(counts_1))
pdf_2 = counts_2/(sum(counts_2))
pdf_3 = counts_3/(sum(counts_3))

#print(pdf);
#print(bin_edges)

#compute CDF
cdf_1 = np.cumsum(pdf_1)
cdf_2 = np.cumsum(pdf_2)
cdf_3 = np.cumsum(pdf_3)

plt.subplot(1,3,1)
plt.plot(bin_edges_1[1:],pdf_1)
plt.plot(bin_edges_1[1:], cdf_1)
plt.title('PDF & CDF of mildly active users')
plt.legend(labels=['PDF','CDF'])

plt.subplot(1,3,2)
plt.plot(bin_edges_2[1:],pdf_2)
plt.plot(bin_edges_2[1:], cdf_2)
plt.title('PDF & CDF of moderately active users')
plt.legend(labels=['PDF','CDF'])

plt.subplot(1,3,3)
plt.plot(bin_edges_3[1:],pdf_3)
plt.plot(bin_edges_3[1:], cdf_3)
plt.title('PDF & CDF of highly active users')
plt.legend(labels=['PDF','CDF'])

plt.show();
```

## **PCA on the Data**


```python
df_1 = df_new.copy()
```

```python
df_1.columns
```

```python
df_1.drop(['Id','ActivityDate','year','month','date','Week','Year','L','M','H','DayOfWeek','zscore_for_TotalDistance'],axis=1,inplace=True)
df_1.head()
```

```python
labels = df_1['ActivityStatus']
data = df_1.drop("ActivityStatus",axis=1)
```

```python
labels.head()
```

```python
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
standardized_data = StandardScaler().fit_transform(data)
print(standardized_data.shape)
```

```python
from sklearn import decomposition
pca = decomposition.PCA()
```

```python
pca.n_components = 2
pca_data = pca.fit_transform(standardized_data)
print(pca_data.shape)
```

```python
pca_data = np.vstack((pca_data.T,labels)).T
```

```python
pca_df = pd.DataFrame(data = pca_data,columns=("feature_1","feature_2","label"))
```

```python
sns.FacetGrid(pca_df,hue='label',height=6).map(plt.scatter,'feature_2','feature_1').add_legend()
plt.show()
```

## **Conclusion**

We arrive at the following conclusions based on the study conducted above:
* The users with high activity status contribute to the majority of the dataset
* It was observed that during the days – Tuesday, Wednesday, and Thursday, the activity among the users was it it’s maximum while the weekends weren’t very productive
* The heart-rate data shows that a majority of users reported normal and healthy heart rates
* The users slept the most on Sundays; overall, the average of the sleep hours is less than the recommended 8 hrs. Hence they must sleep more
* Although the distinction of activity status is based on calories burnt, it is seen that a majority of users (81.5%) have more sedentary minutes. Sedentary minutes should be reduced
* The plot between sedentary minutes and total sleep time shows a negative correlation, hence the amount of sedentary minutes must be practically lowered.
