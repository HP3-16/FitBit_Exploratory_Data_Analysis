# **EXPLORATORY DATA ANALYSIS**


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
df = pd.read_csv('FitBit data.csv')
```

```python
df.isnull().sum()
```

```python
df_new = df.copy()
```

```python
df_new.drop(df_new[df['Calories']==0.0].index,inplace=True)
df_new.describe().T
```

```python
df_new['year'] = pd.DatetimeIndex(df_new['ActivityDate']).year
df_new['month'] = pd.DatetimeIndex(df_new['ActivityDate']).month
df_new['date'] = pd.DatetimeIndex(df_new['ActivityDate']).day
```

```python
df_new['Week'] = pd.to_datetime(df_new.ActivityDate).dt.week
df_new['Year'] = pd.to_datetime(df_new.ActivityDate).dt.year
```

```python
def calorie_class(cal):
  if cal<=1600:
    return "mildly active"
  elif cal>1600 and cal<=2200:
    return "moderately active"
  else:
    return "highly active"
df_new['ActivityStatus'] = df_new['Calories'].apply(calorie_class)
```

```python
df_new['ActivityStatus'].value_counts()
```

```python
df_new.head()
```

```python
df_new.columns
```

```python
df_sample = df_new[['TotalSteps', 'TotalDistance', 'VeryActiveDistance',
       'ModeratelyActiveDistance','Calories',
       'ActivityStatus',]]
sns.set_style("whitegrid");
sns.pairplot(df_sample, hue="ActivityStatus", size=3);
plt.show()
```

## **Univariate Analysis**

```python
df_new.loc[(df_new['ActivityStatus']=='mildly active'),'L'] =1
df_new.loc[(df_new['ActivityStatus']=='moderately active'),'M'] =1
df_new.loc[(df_new['ActivityStatus']=='highly active'),'H'] =1

markers = ['Mild', 'Moderate', 'High']
sizes = [df_new['L'].value_counts()[1],
         df_new['M'].value_counts()[1],
         df_new['H'].value_counts()[1],
         ]
plt.title("The frequency of Activity among the users")
plt.pie(sizes, labels=markers,autopct='%1.1f%%')
plt.show()
```
